### Bivariate statistics

Until now, we've only worked on data that do not depend on other data.  But there are many examples in Earth Science where we care about the dependence between two types of data.  For example, the distance from the ridge crest versus age gives you a spreading rate.  The depth in a sediment core versus age gives you a sedimentation rate.  The ratio of the radioactive form of carbon, $^{14}$C, to a stable form,$^{12}$C, is a function of the age of the material being dated. The difference in arrival times of the $P$ and $S$ seismic waves is related to distance. These examples rely on the use of bivariate statistics to get at the desired quantities.   

### Age of the universe

Let's use as our example data set, the retreat velocity of galaxies as a function of their distance.  This is the basic data set underlying what has come to be known as "Hubble's Law" (same Hubble as for the Hubble telescope). Hubble published his results in 1929 [Hubble, E. P. (1929) Proc. Natl. Acad. Sci., 15, 168â€“173.]  Back then it was not clear whether the universe was static, expanding, or collapsing.  No one knew!  So Hubble hypothesized that if the universe were expanding, then everything in it must be moving away from us. The greater the distance, the faster they must be moving.  So "all" he had to do was measure the distance and velocity of distant galaxies.  

To measure velocity, Hubble made use of something called  the doppler shift. To understand how this works, recall that the pitch you hear as an ambulance approaches changes. During doppler shift, the ambulance's pitch changes from high (as it approaches) to low (as it recedes). The pitch changes  because the relative frequency of the sound waves changes. The frequency increases as the ambulance approaches because the sound waves 'bunch up', leading to a higher pitch, and then decreases as it moves away (and the sound waves spread out), resulting in a lower pitch.  

```python
Image(filename='Figures/doppler.jpg',width=600)
```

_[Figure from http://www.physicsclassroom.com/class/waves/Lesson-3/The-Doppler-Effect]_

 
The same principle applies to light.  Stars are made mainly of hydrogen- remember our class on solar elemental abundances? Hydrogen absorbs energy at discrete wavelengths.  So if light passes through hydrogen, it loses energy at those specific wavelengths.  If a star or galaxy is moving away from us, these distinct absorption bands are shifted towards longer wavelengths or toward the red end of the rainbow spectrum. The faster the star or galaxy travels, the more shifted the bands are: 


```python
Image(filename='Figures/dopp-redshift.jpg',width=600)
```

_[Figure from http://www.a-levelphysicstutor.com/wav-doppler.php]_

So, Hubble measured the red shift of different galaxies and converted them to velocities.  
He then estimated the distance to these objects, which is harder to do.  There are many ways to do this and you can look them up online if you like.  In any case, we can find the dataset of distance to different galaxies and their velocities here: http://keyah.asu.edu/lessons/AgeOfUniverse/Exercises.html

Okay, now let's read in the data and take a look.


```python
data=pd.read_csv('Datasets/redShiftDistance/d_v.csv',header=1)
print (data.head()) # take a peek
plt.plot(data['d (10^3 Mpc)'],data['v (10^3km/s)'],'r*') # plot the points as red stars
plt.ylabel('v (10$^3$km s$^{-1}$)') # notice the latex style labels
plt.xlabel('Distance (10$^3$ Mpc)');
```

So, we have distance on the X axis in Mega Parsecs and velocity on the Y axis in 10$^3$km/s.  

To calculate the age of the universe, we can use Hubble's logic:  

Here is "Hubble's Law":  $v = H_o d$, where $v$ is velocity, $d$ is distance, and $H_o$ is "Hubble's constant".  

This looks a lot like the equation for a line through the data ($y=mx + b$) where $m$ is the slope and $b$ is the y-intercept.  

We can calculate the coefficients $m$ and $b$  in **NumPy** fairly easily, using a function called **np.polyfit( )**.



```python
help(np.polyfit)
```

**np.polyfit( )** can be used to calculate best fit lines (setting the degree (**deg**) to 1), or higher order curves (setting degree to 2 or higher).  

All we want is a best fit line for now, so let's see how well it does:  

```python
print (np.polyfit(data['d (10^3 Mpc)'],data['v (10^3km/s)'],1))
```

Ahah!  So $H_o$, the slope of the best-fit line,  is 70.3 (in some weird units).  

Before going on, let's plot the best fit line on our graph.  

We can assign the best fitting slope and y-intercept from **np.polyfit( )** to a variable (**m_b**) and feed that to another function **np.polyval( )** which will calculate new Y values using the model of a linear fit: 

```python
help(np.polyval)
```

```python
# save the statistics in an array called m_b
m_b= np.polyfit(data['d (10^3 Mpc)'],data['v (10^3km/s)'],1)
print (m_b) # see if that worked
```

**m_b**  seems to be an array of coefficients, where the first is the slope and the second is the y-intercept.  

Let's feed **m_b** into **np.polyval( )**, along with our X array to get a new set of Y values, assuming a linear fit. Then we can   plot the model data as a black line along with the original Y values. 

```python
modelYs=np.polyval(m_b,data['d (10^3 Mpc)'])
# now plot the data and the best-fit line: 
plt.plot(data['d (10^3 Mpc)'],data['v (10^3km/s)'],'r*')
plt.plot(data['d (10^3 Mpc)'],modelYs,'k-') # plot as red line
plt.ylabel('v (10$^3$km s$^{-1}$)')
plt.xlabel('Distance (10$^3$ Mpc)');
```

It would be handy to also know how well our data fit the model (not everything is linear, after all).  

To do that, we need a different function: **scipy.stats.linregress( )**. 

```python
from scipy.stats import stats as stats # import the module
help(stats.linregress)
```

And use it, to get what is normally called the $R^2$ value, which when 1. represents perfect agreement.  

```python
print (stats.linregress(data['d (10^3 Mpc)'],data['v (10^3km/s)']))
# and save the statistics as: 
slope,intercept, r_value, p_value, std_err = stats.linregress(data['d (10^3 Mpc)'],data['v (10^3km/s)'])
print ('\n')
print ('slope: %7.3f, intercept: %4.1f, R^2: %5.3f'%\
    (slope, intercept, r_value**2))
```

Not a bad fit!   

Now let's get back to Hubble's Law and the age of the universe.  

We had $v=H_o d$ as Hubble's law and we know that distance = velocity times time, or,  $d=vt$.  So, if we divide both sides by $v$ and  we get: 

1=$H_o$t. 

Solving for $t$ (the age of the universe), we get 

$t=1/H_o$.  





```python
print (1./slope)
t=1./slope
```

But the units are weird (not years,  per Mpc s/km).  To fix this, we need to know how many kilometers are in a mega parsec.  In fact there are 3.09 x 10$^{19}$ km/Mpc.  
So, the age of the universe in seconds (**Age_sec**)  can be gotten by converting the parsecs to km:

Age (s) = $t \frac{s Mpc}{km}$ x $3.09 x 10^{19} \frac {km}{Mpc}$


```python
Age_sec=t*3.09e19
print (Age_sec)
```

That's a lot of seconds!  We should convert seconds to years.  As it happens, there are approximately   $\pi$x 10$^7$ seconds in a year. 

 

```python
Age_yrs=Age_sec/(np.pi*1e7)
print (Age_yrs)
```

And now in billions of years please:  

```python
print ('Age of the universe: %5.2f'%(Age_yrs*1e-9), 'Ga')
```

### When data are curved. 

There are lots of other curve fitting methods in **scipy** and **NumPy**.  I encourage you to look around when the need arises.  

As an example of a curved relationship, there are new data sets available for the classic Hubble problem.  I found one published by Betoule et al. in 2014 (http://dx.doi.org/10.1051/0004-6361/201423413).   Cosmologists plot things differently now.  They use the parameters $z$ and $\mu$ which are related to the red shift velocity and distance.  $z$ is the fractional shift in the spectral wavelength and $\mu$ has something to do with distance. 

Here is a plot from the Betoule et al. paper: 

```python
Image(filename='Figures/betoule14.png',width=600)
```

_[Figure modified from Betoule et al., 2014.]  The different colors are different types of objects (galaxies, supernovae, etc.)._  

Notice that they sneekily plotted things on a log plot (which hides some surprising things.)

To compare the old estimate of the age of the universe (~13 Ga) with that gotten from the new data set, we need to convert $z$ and $\mu$ to distance and velocity.  


According to http://hyperphysics.phy-astr.gsu.edu/hbase/Astro/hubble.html

velocity $v$ (as fraction of the speed of light, $c$) is given by

${v\over c}= \bigl({{(z+1)^2-1}  \over {(z+1)^2+1}}\bigr)$

where $c=3 *10^8$m s$^{-1}$.  

And according to the Betoule et al. (2014) paper, $\mu$ relates to distance in parsecs $d$ like this:  



$\mu=5\log(d/10)$.   

Let's read in the data (available from this website:  http://cdsarc.u-strasbg.fr/viz-bin/qcat?J/A+A/568/A22#sRM2.2), which are averages of the data shown in the figure above,and take a peek.  




```python
new_data=pd.read_csv('Datasets/redShiftDistance/mu_z.csv',header=1)
#new_data=pd.read_csv('Datasets/redShiftDistance/bigBang.csv')
print (new_data.columns)
```

 
Now we can plot it the same way as the cosmologists do, using $\mu$ and $\log z$: 

```python
plt.plot(np.log10(new_data.z),new_data.mu,'bo')
plt.xlabel('$\log_{10}$ (z) ')
plt.ylabel('$\mu$')
plt.xlim(-2,0);
```

To compare the new data with the 'old', we must .  
- We must transform $z$  to velocity  
- and transform  $\mu$ to distance using the equations provided. 
- And the data go to much farther distances now than the 'old' data set, so we can truncate the data using Pandas tricks. 
- Oh and we can recalculate the age of the universe, making a nice
lambda function to help.

Let's make the function first:  

```python
# function returns age in Ga for Ho
age_from_Ho= lambda Ho : 1e-9*3.09e19/(Ho*np.pi*1e7)
```

Now we do the conversions, truncation and recalculate H$_o$:  

```python
# convert z to velocity in 10^3 km/s (or 10^6 m/s)
c=3e8 # speed of light in m/s
new_data['velocity']=1e-6*c*\
    (((new_data.z+1.)**2-1.)/((new_data.z+1.)**2+1.)) 
# convert mu to distance in 10^3 Mpc (a Gpc):
new_data['distance']=10.*(10.**((new_data['mu'])/5.))*1e-9 # to Gpc
# and filter for the closest objects
close_data=new_data[new_data.distance<0.7]

close_fit= np.polyfit(close_data['distance'],close_data['velocity'],1) # calculate the coefficients
close_modelYs=np.polyval(close_fit,close_data['distance']) # get the model values

age=age_from_Ho(close_fit[0]) # get a new age 
print (age)

```

Wow - that is different!   


Now we can plot it up and see what happened! 

```python
# plot the old data (in red dots) with the new (in blue dots)
plt.plot(data['d (10^3 Mpc)'],data['v (10^3km/s)'],'r*',label='Old data') # as red stars
plt.plot(data['d (10^3 Mpc)'],modelYs,'r--',label='Old linear fit') # plot old fit
plt.plot(close_data.distance,close_data.velocity,'bd',label='New data') # as blue diamonds
plt.plot(close_data.distance,close_modelYs,'b-',label='New Linear fit') # plot old fit
plt.xlabel('Distance (10$^3$ Mpc)')
plt.ylabel('v (10$^3$km s$^{-1}$)')
plt.text(0.5,10,'New Age: %4.1f'%age+' Ga') # put on a note with the new age
plt.legend(numpoints=1,loc=2);  # and a legend
```

It looks like the velocity  from the new data is slower than the velocity estimated in the old data. The age of the universe just got a lot older! But don't get too excited yet!

Although the two data sets look superficially similar, if you read the paper, you will see there are many adjustments required to get the distances right (Hubble was in fact WAY off, like by a factor of 10!).  


We only looked at the data that were 'close', less than .7 Mpc), but that only accounts for a small portion of the data set.  Let's look at the entire dataset.  

```python
plt.plot(new_data.distance,new_data.velocity,'bd',label='new data')
plt.ylabel('v (10$^3$km s$^{-1}$)')
plt.xlabel('Distance (10$^3$ Mpc)');
```

Oops.   That doesn't look linear at all! The cosmologists believe that the expansion of the universe is accelerating.

So maybe we should calculate the fit with a higher order polynomial instead of with a linear fit.   

```python
colors  = ['m','m','r','g','k'] # make a list of colors for plotting
plt.plot(new_data.distance,new_data.velocity,'bd',label='new data')

# try with higher degree polynomials
for deg in range(2,5):
    fit= np.polyfit(new_data['distance'],new_data['velocity'],deg) # second order
    Y=np.polyval(fit,new_data['distance'])
    plt.plot(new_data.distance,Y,colors[deg],label="%i degree"%deg)
plt.ylabel('v (10$^3$km s$^{-1}$)')
plt.xlabel('Distance (10$^3$ Mpc)')
plt.legend(numpoints=1,loc=2);
```

So, it looks like a 3rd degree polynomial does the trick.  But this is really an active area of research (see article in the New York times here: 
https://nyti.ms/2loeFoX).  The apparent acceleration may be controlled by _dark energy_ and values for H$_o$ hover between 67 (~14.7 Ga) and 73 (13.5 Ga).

Certain isotopes are unstable and decay through _radioactive decay_.  The formula for radioactive decay is:

$N= N_o \exp^{-\lambda T}$

  $\lambda$ is the decay constant (the time for $N$ to decay to $1/\exp$ of the original value

  $T$ is time

  $N$ is the number of nuclei remaining after time $T$ 

  $N_o$ is the original number or parent nuclei

The _half-life_ ($t_{1/2}$) is the time for $N$ to decay to $N_o$/2. The formula is:

$t_{1/2} = \frac {ln 2}{\lambda}$
